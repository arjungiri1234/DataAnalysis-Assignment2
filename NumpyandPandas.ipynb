## Problem 3: Vectorized computation (10 points)

Let x be a length-100 NumPy array of random values drawn from a standard normal distribution
(use a fixed seed).
1. Compute the z-scores of x: z = (x − ¯x)/s using vectorized operations.
2. Verify (numerically) that the mean of z is approximately 0 and the standard deviation is ap-
proximately 1.
3. Compute the proportion of entries of z that fall in the interval [−1, 1].
State the seed you used.

```python
# Set seed for reproducibility
np.random.seed(42)
print(f"Seed used: 42")

# Generate length-100 array from standard normal distribution
x = np.random.randn(100)

# 1. Compute z-scores using vectorized operations
x_mean = x.mean()
x_std = x.std()
z = (x - x_mean) / x_std

print(f"\n1. Z-scores computed using: z = (x - mean) / std")
print(f"   First 10 z-scores: {z[:10]}")

# 2. 
z_mean = z.mean()
z_std = z.std()

print(f"\n2. Verification:")
print(f"   Mean of z: {z_mean:.10f} (approximately 0)")
print(f"   Standard deviation of z: {z_std:.10f} (approximately 1)")

# 3. Compute proportion in interval [-1, 1]
in_interval = np.logical_and(z >= -1, z <= 1)
proportion = in_interval.sum() / len(z)

print(f"\n3. Proportion of z in [-1, 1]: {proportion:.4f} ({proportion*100:.2f}%)")
print(f"   Number of values in [-1, 1]: {in_interval.sum()} out of {len(z)}")
```

Create an array M with shape (6, 4) containing random integers from 0 to 9 (use a fixed seed).
1. Compute the sum of each row and store it as a 1D array row sums.

3. Find the index (row, column) of the maximum value in M.
4. Standardize each column of M to have mean 0 and standard deviation 1 (column-wise standard-
ization). Report the mean and std of each standardized column to confirm.

```python
# Set seed for reproducibility
np.random.seed(123)
print(f"Seed used: 123")

# Create 6x4 array with random integers from 0 to 9
M = np.random.randint(0, 10, size=(6, 4))
print(f"\nArray M (shape {M.shape}):")
print(M)

# 1. Compute sum of each row
row_sums = M.sum(axis=1)
print(f"\n1. Row sums (sum along axis=1):")
print(row_sums)
print(f"   Shape: {row_sums.shape}")

# 2. Compute mean of each column
col_means = M.mean(axis=0)
print(f"\n2. Column means (mean along axis=0):")
print(col_means)
print(f"   Shape: {col_means.shape}")

# 3. Find index (row, column) of maximum value
max_value = M.max()
max_index = np.unravel_index(M.argmax(), M.shape)
print(f"\n3. Maximum value in M: {max_value}")
print(f"   Index (row, column): {max_index}")
print(f"   Verification: M[{max_index[0]}, {max_index[1]}] = {M[max_index[0], max_index[1]]}")

# 4. Standardize each column (mean=0, std=1)
M_standardized = (M - M.mean(axis=0)) / M.std(axis=0)
print(f"\n4. Column-wise standardization:")
print(f"   Standardized array M:")
print(M_standardized)

# Verify standardization
standardized_means = M_standardized.mean(axis=0)
standardized_stds = M_standardized.std(axis=0)
print(f"\n   Verification - Mean of each standardized column:")
print(f"   {standardized_means}")
print(f"\n   Verification - Std of each standardized column:")
print(f"   {standardized_stds}")
```

## Problem 5: Linear algebra basics (10 points)

1. Compute Av using NumPy.
2. Compute the determinant of A.
3. Compute the eigenvalues of A (numerically).
4. Solve Ax = v for x.
Briefly explain (1–2 sentences) when you would use np.linalg.solve instead of explicitly com-
puting A^-1



```python
# Define matrix A and vector v
A = np.array([[2, 1, 0],
              [1, 3, 1],
              [0, 1, 2]])

v = np.array([[1],
              [2],
              [3]])

print("Matrix A:")
print(A)
print(f"\nVector v:")
print(v)

# 1. Compute Av using NumPy
Av = A @ v  # or np.dot(A, v) or A.dot(v)
print(f"\n1. Av (matrix-vector multiplication):")
print(Av)

# 2. Compute the determinant of A
det_A = np.linalg.det(A)
print(f"\n2. Determinant of A: {det_A}")

# 3. Compute eigenvalues of A (numerically)
eigenvalues = np.linalg.eigvals(A)
print(f"\n3. Eigenvalues of A:")
print(eigenvalues)

# 4. Solve Ax = v for x
x = np.linalg.solve(A, v)
print(f"\n4. Solution to Av = v  for x:")
print(x)
```

## Brief Explanation
We should use np.linalog.solve instead of explicitly computing A⁻¹ because it is more numerically stable and have efficient computation. The solve function uses algorithms like LU decomposition that are faster and less prone to rounding errors than explicitly inverting a matrix.

## Problem 13: Efficient CSV reading for larger files (10 points)
This problem mirrors the Lecture 6 notebook strategies: reduce memory use and compute results
without loading everything at once.
1. Create (or reuse) a CSV file named sensor data large.csv with at least 200,000 rows and
the columns: timestamp, sensor id, temperature, humidity, pressure, voltage, status,
location, reading quality, battery pct. (You may generate synthetic data similarly to the
lecture notebook.)
2. Read only the columns sensor id and temperature using usecols. Report the DataFrame’s
memory usage (in MB).
3. Using pd.read csv(..., chunksize=...) compute the average temperature per sensor without
loading the full dataset at once. Return a DataFrame with columns sensor id and avg temperature,
sorted by sensor id.
4. Briefly explain (2–3 sentences) why chunking can be helpful even when you could load the entire
file.

```python
# 1. Create synthetic sensor data with 200,000 rows
print("\n1. Creating sensor_data_large.csv with 200,000 rows...")

np.random.seed(42)
n_rows = 200000

sensor_data = {
    'timestamp': pd.date_range('2024-01-01', periods=n_rows, freq='1min'),
    'sensor_id': np.random.choice(['S001', 'S002', 'S003', 'S004', 'S005'], n_rows),
    'temperature': np.random.uniform(15, 35, n_rows),
    'humidity': np.random.uniform(30, 90, n_rows),
    'pressure': np.random.uniform(980, 1020, n_rows),
    'voltage': np.random.uniform(3.0, 5.0, n_rows),
    'status': np.random.choice(['OK', 'WARNING', 'ERROR'], n_rows),
    'location': np.random.choice(['Building_A', 'Building_B', 'Building_C'], n_rows),
    'reading_quality': np.random.uniform(0.5, 1.0, n_rows),
    'battery_pct': np.random.uniform(0, 100, n_rows)
}

sensor_df = pd.DataFrame(sensor_data)
sensor_df.to_csv('sensor_data_large.csv', index=False)

print(f"   Created file with {n_rows:,} rows and {len(sensor_df.columns)} columns")

# 2.Read only the columns sensor id and temperature using usecols. Report the DataFrame’s memory usage (in MB).
df_subset = pd.read_csv('sensor_data_large.csv', usecols=['sensor_id', 'temperature'])

memory_mb = df_subset.memory_usage(deep=True).sum() / (1024 * 1024)

print(f"\n2. Read only 'sensor_id' and 'temperature' columns:")
print(f"   Shape: {df_subset.shape}")
print(f"   Memory usage: {memory_mb:.2f} MB")

# 3. Using pd.read csv(..., chunksize=...) compute the average temperature per sensor without
# loading the full dataset at once. Return a DataFrame with columns sensor id and avg temperature,sorted by sensor id
print("\n3. Computing average temperature per sensor using chunking...")

chunksize = 50000
chunk_results = []

for chunk in pd.read_csv('sensor_data_large.csv', 
                         usecols=['sensor_id', 'temperature'],
                         chunksize=chunksize):
   
    chunk_agg = chunk.groupby('sensor_id').agg({
        'temperature': ['sum', 'count']
    })
    chunk_results.append(chunk_agg)

# Combine all chunks
combined = pd.concat(chunk_results)

# Compute overall average
final_avg = combined.groupby(level=0).sum()
final_avg['avg_temperature'] = final_avg[('temperature', 'sum')] / final_avg[('temperature', 'count')]

# Create final result DataFrame
result = pd.DataFrame({
    'sensor_id': final_avg.index,
    'avg_temperature': final_avg['avg_temperature'].values
}).sort_values('sensor_id').reset_index(drop=True)

print("\n   Average temperature per sensor:")
print(result)
```

# Explanation
Briefly explain (2–3 sentences) why chunking can be helpful even when you could load the entire
file.


Chunking can be helpful even when we could load the entire file because it reduces memory usage by processing datas in smaller chunks and allows processing of datasets that might grow beyound available RAM.

## Problem 15: Auto-loader function (10 points)
Create a function load orders(path) that:
• Detects the file extension (.csv or .json),
• Loads the file into a DataFrame with appropriate options (e.g., parse dates),
• Raises a clear ValueError for unsupported file types.
Provide short examples showing it working for both supported formats.

```python
def load_orders(path):
    from pathlib import Path
    
    # Convert to Path object
    file_path = Path(path)
    
    # Get file extension
    extension = file_path.suffix.lower()
    
    # detects the file extension .csv or json
    if extension == '.csv':
        df = pd.read_csv(file_path, parse_dates=['date'])
        return df
    elif extension == '.json':
        df = pd.read_json(file_path)
        # Ensure date column is datetime if present
        if 'date' in df.columns:
            df['date'] = pd.to_datetime(df['date'])
        return df
    else:
        raise ValueError(f"Unsupported file type: {extension}. Only .csv and .json are supported.")


# Example 1: Load CSV file
try:
    df_csv_loaded = load_orders('orders.csv')
    print(f"Successfully loaded CSV file")
    print(f"Shape: {df_csv_loaded.shape}")
    print(f"Date dtype: {df_csv_loaded['date'].dtype}")
    print("\nFirst 3 rows:")
    print(df_csv_loaded.head(3))
except Exception as e:
    print(f"Error: {e}")

# Example 2: Load JSON file
print("\n" + "-"*50)
print("Example 2: Loading JSON file")
print("-"*50)

try:
    df_json_loaded = load_orders('orders_records.json')
    print(f"Successfully loaded JSON file")
    print(f"Shape: {df_json_loaded.shape}")
    print(f"Date dtype: {df_json_loaded['date'].dtype}")
    print("\nFirst 3 rows:")
    print(df_json_loaded.head(3))
except Exception as e:
    print(f"Error: {e}")

# Example 3: Unsupported file type (demonstrates error handling)
print("\n" + "-"*50)
print("Example 3: Unsupported file type (should raise error)")
print("-"*50)

try:
    df_bad = load_orders('orders.xlsx')
except ValueError as e:
    print(f"Caught expected error: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")
```
