# Problem 1: Array creation and dtypes(10 points)


Create the following arrays:
1. A 1D array a containing the integers 0 through 19.
2. A 2D array b with shape (4, 5) containing the same values as a in row-major order.
3. A float array c that is the result of dividing b by 3.
Report the shape and dtype of each array and briefly explain why c has its dtype.

```python
import numpy as np

# 1) 1D array a: integers 0 through 19
a = np.arange(20)

# 2) 2D array b: shape (4, 5) using row-major order
b = a.reshape(4, 5)

# 3) float array c: divide b by 3
c = b / 3

print("a:\n", a)
print("a shape:", a.shape, "| a dtype:", a.dtype, "\n")

print("b:\n", b)
print("b shape:", b.shape, "| b dtype:", b.dtype, "\n")

print("c:\n", c)
print("c shape:", c.shape, "| c dtype:", c.dtype)
```

# Brief Explnation


c is a float array because integer values 3 uses true division, which can produce non-integer results. Numpy protects the dtype from integer to floating point to represent those decimals.

## Problem 2: Indexing, slicing, and boolean masks (10 points)

Using b from Problem 1:
1. Extract the last column of b.
2. Extract the sub-matrix consisting of rows 1–2 and columns 2–4.
3. Create a boolean mask for entries of b that are divisible by 3, and use it to return those values
as a 1D array.
4. Replace all values in b that are greater than 12 with -1 (without using Python loops).

```python
# 1. Extract the last column of b
last_column = b[:, -1]
print("1. Last column of b:")
print(last_column)
print()

# 2. Extract sub-matrix (rows 1-2, columns 2-4)
sub_matrix = b[1:3, 2:5]
print("2. Sub-matrix (rows 1-2, columns 2-4):")
print(sub_matrix)
print()

# 3. Boolean mask for entries divisible by 3
mask = (b % 3 == 0)
values_divisible_by_3 = b[mask]
print("3. Boolean mask (divisible by 3):")
print(mask)
print("\nValues divisible by 3:")
print(values_divisible_by_3)
print()

# 4. Replace values greater than 12 with -1
b_modified = b.copy()
b_modified[b_modified > 12] = -1
print("4. Array b with values > 12 replaced by -1:")
print(b_modified)
```

## Problem 3: Vectorized computation (10 points)

Let x be a length-100 NumPy array of random values drawn from a standard normal distribution
(use a fixed seed).
1. Compute the z-scores of x: z = (x − ¯x)/s using vectorized operations.
2. Verify (numerically) that the mean of z is approximately 0 and the standard deviation is ap-
proximately 1.
3. Compute the proportion of entries of z that fall in the interval [−1, 1].
State the seed you used.

```python
# Set seed for reproducibility
np.random.seed(42)
print(f"Seed used: 42")

# Generate length-100 array from standard normal distribution
x = np.random.randn(100)

# 1. Compute z-scores using vectorized operations
x_mean = x.mean()
x_std = x.std()
z = (x - x_mean) / x_std

print(f"\n1. Z-scores computed using: z = (x - mean) / std")
print(f"   First 10 z-scores: {z[:10]}")

# 2. 
z_mean = z.mean()
z_std = z.std()

print(f"\n2. Verification:")
print(f"   Mean of z: {z_mean:.10f} (approximately 0)")
print(f"   Standard deviation of z: {z_std:.10f} (approximately 1)")

# 3. Compute proportion in interval [-1, 1]
in_interval = np.logical_and(z >= -1, z <= 1)
proportion = in_interval.sum() / len(z)

print(f"\n3. Proportion of z in [-1, 1]: {proportion:.4f} ({proportion*100:.2f}%)")
print(f"   Number of values in [-1, 1]: {in_interval.sum()} out of {len(z)}")
```

## Problem 4: Aggregation and axis reasoning (10 points)

Create an array M with shape (6, 4) containing random integers from 0 to 9 (use a fixed seed).
1. Compute the sum of each row and store it as a 1D array row sums.
2. Compute the mean of each column and store it as a 1D array col means.
3. Find the index (row, column) of the maximum value in M.
4. Standardize each column of M to have mean 0 and standard deviation 1 (column-wise standard-
ization). Report the mean and std of each standardized column to confirm.

```python
# Set seed for reproducibility
np.random.seed(123)
print(f"Seed used: 123")

# Create 6x4 array with random integers from 0 to 9
M = np.random.randint(0, 10, size=(6, 4))
print(f"\nArray M (shape {M.shape}):")
print(M)

# 1. Compute sum of each row
row_sums = M.sum(axis=1)
print(f"\n1. Row sums (sum along axis=1):")
print(row_sums)
print(f"   Shape: {row_sums.shape}")

# 2. Compute mean of each column
col_means = M.mean(axis=0)
print(f"\n2. Column means (mean along axis=0):")
print(col_means)
print(f"   Shape: {col_means.shape}")

# 3. Find index (row, column) of maximum value
max_value = M.max()
max_index = np.unravel_index(M.argmax(), M.shape)
print(f"\n3. Maximum value in M: {max_value}")
print(f"   Index (row, column): {max_index}")
print(f"   Verification: M[{max_index[0]}, {max_index[1]}] = {M[max_index[0], max_index[1]]}")

# 4. Standardize each column (mean=0, std=1)
M_standardized = (M - M.mean(axis=0)) / M.std(axis=0)
print(f"\n4. Column-wise standardization:")
print(f"   Standardized array M:")
print(M_standardized)

# Verify standardization
standardized_means = M_standardized.mean(axis=0)
standardized_stds = M_standardized.std(axis=0)
print(f"\n   Verification - Mean of each standardized column:")
print(f"   {standardized_means}")
print(f"\n   Verification - Std of each standardized column:")
print(f"   {standardized_stds}")
```
