# Problem 1: Array creation and dtypes(10 points)


Create the following arrays:
1. A 1D array a containing the integers 0 through 19.
2. A 2D array b with shape (4, 5) containing the same values as a in row-major order.
3. A float array c that is the result of dividing b by 3.
Report the shape and dtype of each array and briefly explain why c has its dtype.

```python
import numpy as np

# 1) 1D array a: integers 0 through 19
a = np.arange(20)

# 2) 2D array b: shape (4, 5) using row-major order
b = a.reshape(4, 5)

# 3) float array c: divide b by 3
c = b / 3

print("a:\n", a)
print("a shape:", a.shape, "| a dtype:", a.dtype, "\n")

print("b:\n", b)
print("b shape:", b.shape, "| b dtype:", b.dtype, "\n")

print("c:\n", c)
print("c shape:", c.shape, "| c dtype:", c.dtype)
```

## Problem 3: Vectorized computation (10 points)

Let x be a length-100 NumPy array of random values drawn from a standard normal distribution
(use a fixed seed).
1. Compute the z-scores of x: z = (x − ¯x)/s using vectorized operations.
2. Verify (numerically) that the mean of z is approximately 0 and the standard deviation is ap-
proximately 1.
3. Compute the proportion of entries of z that fall in the interval [−1, 1].
State the seed you used.

```python
# Set seed for reproducibility
np.random.seed(42)
print(f"Seed used: 42")

# Generate length-100 array from standard normal distribution
x = np.random.randn(100)

# 1. Compute z-scores using vectorized operations
x_mean = x.mean()
x_std = x.std()
z = (x - x_mean) / x_std

print(f"\n1. Z-scores computed using: z = (x - mean) / std")
print(f"   First 10 z-scores: {z[:10]}")

# 2. 
z_mean = z.mean()
z_std = z.std()

print(f"\n2. Verification:")
print(f"   Mean of z: {z_mean:.10f} (approximately 0)")
print(f"   Standard deviation of z: {z_std:.10f} (approximately 1)")

# 3. Compute proportion in interval [-1, 1]
in_interval = np.logical_and(z >= -1, z <= 1)
proportion = in_interval.sum() / len(z)

print(f"\n3. Proportion of z in [-1, 1]: {proportion:.4f} ({proportion*100:.2f}%)")
print(f"   Number of values in [-1, 1]: {in_interval.sum()} out of {len(z)}")
```

## Problem 4: Aggregation and axis reasoning (10 points)

Create an array M with shape (6, 4) containing random integers from 0 to 9 (use a fixed seed).
1. Compute the sum of each row and store it as a 1D array row sums.
2. Compute the mean of each column and store it as a 1D array col means.
3. Find the index (row, column) of the maximum value in M.
4. Standardize each column of M to have mean 0 and standard deviation 1 (column-wise standard-
ization). Report the mean and std of each standardized column to confirm.

```python
# Set seed for reproducibility
np.random.seed(123)
print(f"Seed used: 123")

# Create 6x4 array with random integers from 0 to 9
M = np.random.randint(0, 10, size=(6, 4))
print(f"\nArray M (shape {M.shape}):")
print(M)

# 1. Compute sum of each row
row_sums = M.sum(axis=1)
print(f"\n1. Row sums (sum along axis=1):")
print(row_sums)
print(f"   Shape: {row_sums.shape}")

# 2. Compute mean of each column
col_means = M.mean(axis=0)
print(f"\n2. Column means (mean along axis=0):")
print(col_means)
print(f"   Shape: {col_means.shape}")

# 3. Find index (row, column) of maximum value
max_value = M.max()
max_index = np.unravel_index(M.argmax(), M.shape)
print(f"\n3. Maximum value in M: {max_value}")
print(f"   Index (row, column): {max_index}")
print(f"   Verification: M[{max_index[0]}, {max_index[1]}] = {M[max_index[0], max_index[1]]}")

# 4. Standardize each column (mean=0, std=1)
M_standardized = (M - M.mean(axis=0)) / M.std(axis=0)
print(f"\n4. Column-wise standardization:")
print(f"   Standardized array M:")
print(M_standardized)

# Verify standardization
standardized_means = M_standardized.mean(axis=0)
standardized_stds = M_standardized.std(axis=0)
print(f"\n   Verification - Mean of each standardized column:")
print(f"   {standardized_means}")
print(f"\n   Verification - Std of each standardized column:")
print(f"   {standardized_stds}")
```

## Problem 5: Linear algebra basics (10 points)

1. Compute Av using NumPy.
2. Compute the determinant of A.
3. Compute the eigenvalues of A (numerically).
4. Solve Ax = v for x.
Briefly explain (1–2 sentences) when you would use np.linalg.solve instead of explicitly com-
puting A^-1



```python
# Define matrix A and vector v
A = np.array([[2, 1, 0],
              [1, 3, 1],
              [0, 1, 2]])

v = np.array([[1],
              [2],
              [3]])

print("Matrix A:")
print(A)
print(f"\nVector v:")
print(v)

# 1. Compute Av using NumPy
Av = A @ v  # or np.dot(A, v) or A.dot(v)
print(f"\n1. Av (matrix-vector multiplication):")
print(Av)

# 2. Compute the determinant of A
det_A = np.linalg.det(A)
print(f"\n2. Determinant of A: {det_A}")

# 3. Compute eigenvalues of A (numerically)
eigenvalues = np.linalg.eigvals(A)
print(f"\n3. Eigenvalues of A:")
print(eigenvalues)

# 4. Solve Ax = v for x
x = np.linalg.solve(A, v)
print(f"\n4. Solution to Av = v  for x:")
print(x)
```

## Brief Explanation
We should use np.linalog.solve instead of explicitly computing A⁻¹ because it is more numerically stable and have efficient computation. The solve function uses algorithms like LU decomposition that are faster and less prone to rounding errors than explicitly inverting a matrix.

## Problem 8: GroupBy aggregations (10 points)
Compute a summary table by city with:
• total revenue,
• total units,
• number of distinct customers,
• average unit price (simple mean of unit price).
Sort cities by total revenue (descending).

```python
# Group by city and compute aggregations
city_summary = df.groupby('city')

# Calculate each metric separately
total_revenue = city_summary['revenue'].sum()
total_units = city_summary['units'].sum()
distinct_customers = city_summary['customer'].nunique()
avg_unit_price = city_summary['unit_price'].mean()

# Combine into a DataFrame
summary = pd.DataFrame({
    'total_revenue': total_revenue,
    'total_units': total_units,
    'distinct_customers': distinct_customers,
    'avg_unit_price': avg_unit_price
})

# Sort by total revenue descending
summary = summary.sort_values('total_revenue', ascending=False)

print("\nSummary table by city (sorted by total revenue descending):")
print(summary)
```

## Problem 11: Missing data (8 points)
Make a copy of df called df2 and then:
1. Set unit price to missing (NaN) for order id == 106.
2. Show how many missing values each column has.
3. Replace missing unit price with the median unit price within the same category. (Hint:
groupby + transform.)
4. Recompute revenue for df2 and verify order id 106 now has a non-missing revenue.

```python
# Make a copy of df
df2 = df.copy()

# 1. Set unit_price to NaN for order_id == 106
df2.loc[df2['order_id'] == 106, 'unit_price'] = np.nan

print("\n1. Set unit_price to NaN for order_id == 106:")
print(df2[df2['order_id'] == 106])

# 2. Show how many missing values each column has
missing_counts = df2.isnull().sum()

print("\n2. Missing values per column:")
print(missing_counts)

# 3. Replace missing unit_price with median within same category
df2['unit_price'] = df2.groupby('category')['unit_price'].transform(lambda x: x.fillna(x.median()))

print("\n3. After filling missing unit_price with category median:")
print(df2[df2['order_id'] == 106])

# 4. Recompute revenue for df2
df2['revenue'] = df2['units'] * df2['unit_price']

print("\n4. Recomputed revenue for df2:")
print(df2[df2['order_id'] == 106])

# Verify order_id 106 now has non-missing revenue
print(f"\n   Verification:")
print(f"   Order 106 revenue is NaN: {df2[df2['order_id'] == 106]['revenue'].isna().values[0]}")
print(f"   Order 106 revenue value: {df2[df2['order_id'] == 106]['revenue'].values[0]:.2f}")
```
